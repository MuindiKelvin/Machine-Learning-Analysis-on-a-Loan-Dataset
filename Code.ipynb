<h2 style="color:blue">Data Preparation and Processing</h2>

```python
# Split data into features and target
X = df.drop('loan_status', axis=1)  # Features
y = df['loan_status']  # Target variable (Accepted/Rejected)
```

```python
# Encode categorical variables
categorical_cols = [col for col in X.columns if X[col].dtype == 'object']
print(f'Categorical columns: {categorical_cols}')
```

```python
# One-hot encode categorical variables
encoder = OneHotEncoder(handle_unknown='ignore')
X_encoded_cat = encoder.fit_transform(X[categorical_cols])
```

```python
# Dropping the original categorical columns
X_encoded = X.drop(categorical_cols, axis=1)

# Combining the encoded columns with the numerical columns
if isinstance(X_encoded, pd.DataFrame):
    X_encoded = sparse.csr_matrix(X_encoded.values)

X_encoded = sparse.hstack([X_encoded, X_encoded_cat])
```

```python
# Splitting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Creating an imputer object
imputer = SimpleImputer(strategy='mean')  # 'median' or 'most_frequent' can also be used

# Fit the imputer on the training data and transform both training and test data
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
```

```python
from sklearn.impute import SimpleImputer

# Instantiating SimpleImputer to replace NaN values with the mean
imputer = SimpleImputer(strategy='mean')

# Fitting and transforming the imputer on the training data
X_train_imputed = imputer.fit_transform(X_train)

# Transforming the test data using the fitted imputer
X_test_imputed = imputer.transform(X_test)
```

<h2 style="color:blue"> Modeling</h2>

```python
# Train and evaluate Logistic Regression model
print('Logistic Regression')
log_reg = LogisticRegression()
log_reg.fit(X_train_imputed, y_train)  # or X_train_clean, y_train_clean
y_pred = log_reg.predict(X_test_imputed)  # or X_test_clean
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

```python
# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python
# Train and evaluate Decision Tree Classifier
print('\nDecision Tree Classifier')
dt_clf = DecisionTreeClassifier()
# Train the decision tree classifier on the imputed training data
dt_clf.fit(X_train_imputed, y_train)
# Make predictions on the imputed test data
y_pred = dt_clf.predict(X_test_imputed)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

```python
# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python
# Train and evaluate Random Forest Classifier
print('\nRandom Forest Classifier')
random_forest = RandomForestClassifier()
# Train the Random Forest classifier on the imputed training data
random_forest.fit(X_train_imputed, y_train)
# Make predictions on the imputed test data
y_pred = random_forest.predict(X_test_imputed)
# Generate classification report
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

```python
# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

<h2 style="color:blue">Comparison of Classification Accuracy</h2>

```python
from sklearn.metrics import accuracy_score

# Train and evaluate Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(X_train_imputed, y_train)
y_pred_log_reg = log_reg.predict(X_test_imputed)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)

# Train and evaluate Decision Tree Classifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train_imputed, y_train)
y_pred_dt_clf = dt_clf.predict(X_test_imputed)
accuracy_dt_clf = accuracy_score(y_test, y_pred_dt_clf)

# Train and evaluate Random Forest Classifier
random_forest = RandomForestClassifier()
random_forest.fit(X_train_imputed, y_train)
y_pred_random_forest = random_forest.predict(X_test_imputed)
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)

# Compare accuracy scores
print('Accuracy Scores:')
print('Logistic Regression:', accuracy_log_reg)
print('Decision Tree Classifier:', accuracy_dt_clf)
print('Random Forest Classifier:', accuracy_random_forest)

# Identify and print the best performing model
best_model = max(accuracy_log_reg, accuracy_dt_clf, accuracy_random_forest)
if best_model == accuracy_log_reg:
    print('Best Performing Model: Logistic Regression')
elif best_model == accuracy_dt_clf:
    print('Best Performing Model: Decision Tree Classifier')
else:
    print('Best Performing Model: Random Forest Classifier')
```
